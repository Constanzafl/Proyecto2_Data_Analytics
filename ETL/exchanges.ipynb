# EXCHANGES
* Me conecto a la API de CoinGecko y selecciono las primeras 20 exchange del ranking

```python
import pandas as pd
import requests

url = "https://api.coingecko.com/api/v3/exchanges"

response = requests.get(url)
exchanges = response.json()

exchange_data = []

for exchange in exchanges:
    exchange_data.append({
        'ID': exchange['id'],
        'Nombre': exchange['name'],
        'año_inicio': exchange["year_established"],
        'Pais': exchange["country"],
        'Score_confianza': exchange["trust_score"],
        'Ranking_sc':exchange["trust_score_rank"],
        'Vol_trans_btc_24hs': exchange["trade_volume_24h_btc"]
    })

df = pd.DataFrame(exchange_data)

```
