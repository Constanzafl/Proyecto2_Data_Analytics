# Eleccion de 10 monedas
* Elijo 10 criptomonedas teniendo como criterio las 10 con mas capitalizacion de mercado. 

```python
import pandas as pd
import datetime as dt
import time as t
#me descargo pycoingecko para conectarme a la API
from pycoingecko import CoinGeckoAPI
```

```python
#Creo una variable para conectarme
cg = CoinGeckoAPI()

#Confirmo que este conectada
cg.ping()
```

```python
#Uso la funcion de get coin list como para ver los id, simbolo y nombre.
coinList = cg.get_coins_list()
coinDataFrame = pd.DataFrame.from_dict(coinList).sort_values('id'
                                      ).reset_index(drop=True)
```

```python
#la transofrmo en Dataframe
coinDataFrame.head()
```

```python
## EN LA PAGINA DE COINGEKO ME FIJO CUALES SON LAS 10 MONEDAS CON MAS CAPITALIZACION DE MERCADO Y ELIJO ESAS
```

```python
# creo un DF con las 10 que elijo el id, el symbolo y el nombre. 
top10= coinDataFrame[(coinDataFrame['name']=='Bitcoin')|(coinDataFrame['name']=='Ethereum')|(coinDataFrame['name']=='Tether')
                     |(coinDataFrame['name']=='BNB')|(coinDataFrame['name']=='XRP')|(coinDataFrame['name']=='USD Coin')
                     |(coinDataFrame['name']=='Lido Staked Ether')|(coinDataFrame['name']=='Cardano')
                     |(coinDataFrame['name']=='Solana')|(coinDataFrame['name']=='Dogecoin')]
top10= top10.reset_index(drop=True)
top10
```

```python
#pondo los ids en lista para crear una variable para colocarla luego en la otra funcion
coins=top10['id'].tolist()
```

```python
coins
```

```python
#Me traigo la moneda que voy a usar que es USD
counterCurrencies = cg.get_supported_vs_currencies()
vsCurrencies = ['usd']
```

```python
#Traigo con la funcion get coins market, el market cap, rank, prices, volume, market data, etc.
#usando la lista de coins que elegi
coinsMktDataByIds = cg.get_coins_markets(vs_currency = 'usd',
                                        ids = coins)
coinsMktDataByIds
```

```python
#ahora lo transformo en DF
DFcoinsmkt= pd.DataFrame(coinsMktDataByIds)
```

```python
#elimino las columnas que no voy a usar
columnas=['symbol','image','fully_diluted_valuation','circulating_supply','total_supply','max_supply','roi', 'last_updated']
dfcoinmkt= DFcoinsmkt.drop(columns=columnas)
dfcoinmkt
```
